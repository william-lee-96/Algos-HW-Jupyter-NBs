```python
# Merge Two Sorted Arrays (Problem 4)

import numpy as np

path = '/Users/williamlee/Desktop/BMI2005_Algorithms_2020/Labs/rosalind_mer.txt'
with open(path, 'r') as file:
    
    file.readline()
    front = np.array([int(x) for x in file.readline().strip().split(' ')])

    file.readline()
    back = np.array([int(x) for x in file.readline().strip().split(' ')])

def merge (front, back):
    f_pos = 0
    b_pos = 0
    
    merge = np.zeros(len(front) + len(back))
    
    for i in range (len(merge)):
        if f_pos == len(front):
            merge[i] = back[b_pos]
            b_pos += 1
            
        elif b_pos == len(back):
            merge[i] = front[f_pos]
            f_pos += 1
            
        elif front[f_pos] < back[b_pos]:
            merge[i] = front[f_pos]
            f_pos += 1
            
        else: 
            merge[i] = back[b_pos]
            b_pos += 1
            
    return merge

soln = merge(front,back)

print(' '.join(map(str,[int(x) for x in list(soln)])))
```

```python
# Merge Sort (Problem 5)

with open('/Users/williamlee/Desktop/BMI2005_Algorithms_2020/Labs/rosalind_ms.txt','r') as file:
    file.readline()
    array_P5 = np.array([int(x) for x in file.readline().strip().split(' ')])

def merge_sort(array):
    lenArray = len(array)
    
    # base case
    if lenArray <= 1:
        return array

    m = int(lenArray/2) 
    
    front = merge_sort(array[0:m])
    back = merge_sort(array[m:])
    
    return merge(front, back)

print(' '.join(map(str,[int(x) for x in list(merge_sort(array_P5))])))
```

```python
# Quick Sort (Problem 6)

def partition(array, low, high): 
    pivotVal = array[low] 
    
    left = low + 1    
    right = high

    complete = False
    
    while not complete:
        while left <= right and array[left] <= pivotVal:
            left = left + 1
        
        while array[right] >= pivotVal and right >= left:
            right = right - 1

        if right < left:
            complete = True
            
        else:
            array[left], array[right] = array[right], array[left]

    array[low], array[right] = array[right], array[low]

    return right, array

with open('/Users/williamlee/Desktop/BMI2005_Algorithms_2020/Labs/rosalind_qs.txt','r') as file:
    file.readline()
    array_P6 = np.array([int(x) for x in file.readline().strip().split(' ')])

def quick_sort(array, low, high):
    if low < high:
        
        split, array = partition(array, low, high)

        array = quick_sort(array, low, split - 1)
        array = quick_sort(array, split + 1, high)  
        
    return array

np.random.shuffle(array_P6)
print(' '.join(map(str,[int(x) for x in list(quick_sort(array_P6,0,len(array_P6)-1))])))
```
